## Read Pre
### Clinical

```r
pre_clinical <- read.csv("Nathanson2017_CTLA4_Melanoma_RNASeq_Pre/ICB.Nathanson2017_Ipilimumab_Melanoma_Pre.clinical", sep="\t")
```

```r
pre_clinical
```

```r
row.names(pre_clinical) <- pre_clinical$Patient
```

```r
pre_clinical <- pre_clinical[, -1]
```

```r
pre_clinical
```

### Self Subtract

```r
pre_self_subtract <- read.csv("Nathanson2017_CTLA4_Melanoma_RNASeq_Pre/ICB.Nathanson2017_Ipilimumab_Melanoma_Pre.self_subtract", sep="\t")
pre_self_subtract
```

```r
pre_self_subtract <- as.data.frame(t(pre_self_subtract))
```

```r
pre_self_subtract.names <- pre_self_subtract[1,]

colnames(pre_self_subtract) <- pre_self_subtract.names
```

```r
pre_self_subtract <- pre_self_subtract[-1,] 
```

```r
pre_self_subtract
```

### Merge clinical and self subtract

```r
pre_combined <- merge(pre_clinical,pre_self_subtract,by='row.names',all=TRUE)
colnames(pre_combined)[colnames(pre_combined) == "Row.names"] <- "Patient"
pre_combined
```
