## Read Naive

### Clinical

```r
naive_clinical <- read.csv("Riaz2017_PD1_Melanoma_RNASeq_Ipi.Naive/ICB.Riaz2017_Nivolumab_Melanoma_Naive.clinical", sep="\t")
naive_clinical
```

```r
row.names(naive_clinical) <- naive_clinical$X
naive_clinical <- naive_clinical[, -1]
naive_clinical
```

### Self Subtract

```r
naive_self_subtract <- read.csv("Riaz2017_PD1_Melanoma_RNASeq_Ipi.Naive/ICB.Riaz2017_Nivolumab_Melanoma_Naive.self_subtract", sep="\t")
naive_self_subtract
```

```r
naive_self_subtract <- as.data.frame(t(naive_self_subtract))
naive_self_subtract.names <- naive_self_subtract[1,]

colnames(naive_self_subtract) <- naive_self_subtract.names
naive_self_subtract <- naive_self_subtract[-1,] 
naive_self_subtract 
```

### Merge

```r
naive_combined <- merge(naive_clinical, naive_self_subtract,by='row.names',all=TRUE)
colnames(naive_combined)[colnames(naive_combined) == "Row.names"] <- "Patient"
naive_combined
```
