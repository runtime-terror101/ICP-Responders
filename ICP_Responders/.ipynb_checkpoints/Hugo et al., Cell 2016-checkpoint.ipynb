## Clinical

```r
clinical <- read.csv("Hugo2016_PD1_Melanoma_RNASeq/ICB.Hugo2016_Pembrolizumab_Melanoma.clinical", sep="\t")
clinical
```

## Self Subtract

```r
self_subtract <- read.csv("Hugo2016_PD1_Melanoma_RNASeq/ICB.Hugo2016_Pembrolizumab_Melanoma.self_subtract", sep="\t")
self_subtract
```

```r
self_subtract <- as.data.frame(t(self_subtract))  # transpose
self_subtract.names <- self_subtract[1,]  # index by patient ID
colnames(self_subtract) <- self_subtract.names
self_subtract <- self_subtract[-1,] 
self_subtract
```

## Merge

```r
combined <- merge(clinical,self_subtract,by='row.names',all=TRUE)
colnames(combined)[colnames(combined) == "Row.names"] <- "Patient"
combined
```

# Writing final tables in file

```r
setwd("/home/jp/ICP_Responders/FinalTables")
getwd()
```

```r
write.csv(combined, "Hugo.csv", row.names = FALSE)
df <- read.csv("Hugo.csv")
df
```
