## Self Subtract

```r
self_subtract <- read.csv("Lauss2017_ACT_Melanoma_RNASeq/ICB.Lauss2017_ACT_Melanoma.self_subtract", sep="\t")
self_subtract
```

```r
self_subtract <- as.data.frame(t(self_subtract))
self_subtract.names <- self_subtract[1,]
colnames(self_subtract) <- self_subtract.names
self_subtract <- self_subtract[-1,] 
self_subtract
```

## Merge

```r
combined <- merge(clinical,self_subtract,by='row.names',all=TRUE)
colnames(combined)[colnames(combined) == "Row.names"] <- "Patient"
combined
```

# Writing final tables in file
